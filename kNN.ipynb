```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import scale
```

```python
df = pd.read_csv('wine.data', names=["Target", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium",
                                    "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", 
                                     "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"])
```

```python
y = df.Target
X = df.drop("Tar
            get", axis=1)
X.describe()
```

```python
folds = KFold(n_splits=5, shuffle=True, random_state=42) 
```

```python
k_max = 0
mean_max = 0
for k in range(1, 51):
    m = KNeighborsClassifier(n_neighbors=k)
    res = cross_val_score(estimator=m, y=y, X=X,  cv=folds)
    if res.mean() > mean_max:
        k_max = k
        mean_max = res.mean()
    print(f"{k}: {res.mean()}")
print ("-" * 50)
print(f"{k_max}: {mean_max}")
with open("knn_1.txt", "w") as f:
    print(k_max, file=f, end='')
with open("knn_2.txt", "w") as f:
    print("%.2f" % (mean_max), file=f, end='')
```
